```python
from smart_uae_agent import agent_init

agent = agent_init()

tests = [
    ("Attractions query", "What are the main attractions in Abu Dhabi?"),
    ("Budget calculation", "How much for a 4-day luxury trip to Dubai?"),
    ("Prayer times", "Prayer times in Sharjah for tomorrow"),
    ("Trip planning", "Give me a 2-day itinerary for Ras Al Khaimah"),
    
    ("Memory test", "What attractions did you mention earlier?")
]

for test_name, question in tests:
    print(f"=== {test_name.upper()} ===")
    print(f"Question: {question}")
    response = agent.invoke({"input": question})
    print(f"Response: {response['output']}\n")
    print("-" * 50)
```
